```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Step 1: Create and save dummy data
np.random.seed(42)
size = 100

data = {
    "X": np.random.rand(size) * 100,  # Independent variable
    "Y": np.random.rand(size) * 50 + 20  # Dependent variable
}

# Introduce some missing values and duplicates
data["X"][5] = np.nan
data["Y"][10] = np.nan
data["X"][20] = data["X"][21]

df = pd.DataFrame(data)
df.to_csv("dummy_data.csv", index=False)

# Step 2: Load and clean data
df = pd.read_csv("dummy_data.csv")
df = df.drop_duplicates()  # Remove duplicate rows
df = df.dropna()  # Remove missing values

# Step 3: Define independent (X) and dependent (Y) variables
X = df[["X"]]
Y = df["Y"]

# Split data into training and testing sets (80% train, 20% test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Step 4: Perform Linear Regression
model = LinearRegression()
model.fit(X_train, Y_train)

# Predictions
Y_pred = model.predict(X_test)

# Step 5: Plot the regression line
plt.figure(figsize=(8, 5))
plt.scatter(X_test, Y_test, color="blue", label="Actual Data")
plt.plot(X_test, Y_pred, color="red", linewidth=2, label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression Plot")
plt.legend()
plt.show()
```
